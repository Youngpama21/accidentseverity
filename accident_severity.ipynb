```python

# convert time strings to seconds
def time_to_seconds(time_str):
    try:
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s
    except (ValueError, AttributeError):
        return 0

#LabelEncoder
le = LabelEncoder()

# Fit the encoder
df['Accident_severity'] = le.fit_transform(df['Accident_severity'])

#convert all object (string) columns to numerical
for column in xtrain.select_dtypes(include=['object']).columns:
    # Check if the column contains time-like values
    if xtrain[column].str.contains(':', regex=False).any():
        # Convert time-like values to seconds
        xtrain[column] = xtrain[column].apply(time_to_seconds)
        xtest[column] = xtest[column].apply(time_to_seconds)  # Apply to xtest as well
    elif xtrain[column].str.contains('day', regex=False).any():
        # Convert day names to numerical representation (e.g., Monday=0, Tuesday=1, etc.)
        xtrain[column] = pd.Categorical(xtrain[column]).codes
        xtest[column] = pd.Categorical(xtest[column], categories=xtrain[column]).codes
    else:
        # Convert the column to string type before Label Encoding
        xtrain[column] = xtrain[column].astype(str)
        xtest[column] = xtest[column].astype(str)

        #Encoding to convert categorical features to numerical
        le = LabelEncoder()
        xtrain[column] = le.fit_transform(xtrain[column])
        xtest[column] = le.transform(xtest[column])

#applying SMOTE
oversample = SMOTE()
xtrain, ytrain = oversample.fit_resample(xtrain, ytrain)
counter = Counter(ytrain)
print(counter)
print('===========================')


```
